```python
#@title Trabalho planilha rpg  combate V1{ vertical-output: true, display-mode: "form" }
class Personagem:
    def __init__(self, nome, classe):
        self.__nome = nome
        self.__classe = classe
        self.__vida = 100
        self.__mana = 50
    
    def __str__(self):
        return f"{self.__nome}, {self.__classe} ({self.__vida} HP, {self.__mana} MP)"
    
    def atacar(self):
        print(f"{self.__nome} atacou com um ataque básico!")
    
    def defender(self):
        print(f"{self.__nome} se defendeu!")
    
    def usar_habilidade(self, habilidade):
        print(f"{self.__nome} usou a habilidade {habilidade}!")
    
    def tomar_dano(self, dano):
        self.__vida -= dano
        if self.__vida <= 0:
            print(f"{self.__nome} foi derrotado!")
    
    def recuperar_vida(self, cura):
        self.__vida = min(self.__vida + cura, 100)
    
    def recuperar_mana(self, mana):
        self.__mana = min(self.__mana + mana, 50)
    
    def get_nome(self):
        return self.__nome
    
    def get_classe(self):
        return self.__classe
    
    def get_vida(self):
        return self.__vida
    
    def get_mana(self):
        return self.__mana


class Guerreiro(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Guerreiro")
    
    def atacar(self):
        print(f"{self.get_nome()} atacou com um golpe de espada!")
    
    def defender(self):
        print(f"{self.get_nome()} bloqueou o ataque com o escudo!")
    
    def usar_habilidade(self, habilidade):
        if habilidade == "Berserk":
            print(f"{self.get_nome()} usou a habilidade Berserk e ficou mais forte!")
            self.recuperar_vida(10)
        elif habilidade == "Grito de Guerra":
            print(f"{self.get_nome()} usou o Grito de Guerra e inspirou seus aliados!")
            self.recuperar_mana(10)
        else:
            super().usar_habilidade(habilidade)


class Mago(Personagem):
    def __init__(self, nome):
        super().__init__(nome, "Mago")
    
    def atacar(self):
        print(f"{self.get_nome()} lançou uma bola de fogo!")
        self.recuperar_mana(5)
    
    def defender(self):
        print(f"{self.get_nome()} conjurou um escudo mágico!")
        self.recuperar_mana(5)
    
    def usar_habilidade(self, habilidade):
        if habilidade == "Teleporte":
            print(f"{self.get_nome()} usou a habilidade Teleporte e desapareceu!")
        elif habilidade == "Cura":
            print(f"{self.get_nome()} usou a habilidade Cura e se curou!")
            self.recuperar_vida(10)
            self.recuperar_mana(10)
        else:
            super().usar_habilidade(habilidade)


# Testando as classes
guerreiro1 = Guerreiro("Conan")
mago1 = Mago("Gandalf")

print(guerreiro1)
print(mago1)

guerreiro1.atacar()
mago1.atacar()

guerreiro1.defender()
mago1.defender()

guerreiro1.usar_habilidade("Berserk")
mago1.usar_habilidade("Cura")

guerreiro1.tomar_dano(50)
mago1.tomar_dano(70)

print(guerreiro1)
print(mago1)

```
